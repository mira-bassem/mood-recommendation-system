```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

data_dir = r"C:\Users\mirab\OneDrive\Desktop\neural\dataset_by_genre"
```

```python
datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    validation_split=0.2
)
```

```python
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)
```

```python
val_generator = datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=True
)
```

```python
print("Classes:", train_generator.class_indices)
print("Train samples:", train_generator.samples)
print("Validation samples:", val_generator.samples)
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import save_img

save_dir = "processed_dataset"
os.makedirs(save_dir, exist_ok=True)

for class_name in train_generator.class_indices.keys():
    os.makedirs(os.path.join(save_dir, class_name), exist_ok=True)

count = 0
for images, labels in train_generator:
    for i in range(len(images)):
        class_idx = np.argmax(labels[i])
        class_name = list(train_generator.class_indices.keys())[class_idx]

        save_path = os.path.join(save_dir, class_name, f"img_{count}.jpg")
        save_img(save_path, images[i])

        count += 1

    if count >= train_generator.samples:
        break

print("Processed images saved.")
```
